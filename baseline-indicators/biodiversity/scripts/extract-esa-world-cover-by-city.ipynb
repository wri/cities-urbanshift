```python
import geemap
import ee 
import os
import geopandas as gpd
import json
import requests
from geemap import geojson_to_ee, ee_to_geojson
from ipyleaflet import GeoJSON
```

```python
Map = geemap.Map()
Map
```

```python
# Get ESA World cover
world_cover_collection = ee.ImageCollection("ESA/WorldCover/v100")
world_cover_image = world_cover_collection.first()
Map.addLayer(world_cover_image, {}, "WorldCover")
```

```python
# convert geojson boundary into ee feature collection
def city_boundary_geojson_to_ee(city_id):
    # read geojson
    geojson_path = './data/boundaries/admin_boundaries_'+city_id+'.geojson'
    file_path = os.path.abspath(geojson_path)
    # open geojson
    with open(file_path) as f:
        json_data = json.load(f)
    # convert geojson to ee feature collection
    ee_data = geojson_to_ee(json_data)
    
    return ee_data
```

# loop on all cities

```python
# get cities ids
boundaries = gpd.read_file('./data/boundaries/administrative_boundaries.geojson')
```

```python
for i in range(len(boundaries)):
    print(i)
    city_id = boundaries.loc[i, 'city_name']
    print("\n City Id: "+city_id)
    city_boundary_ee = city_boundary_geojson_to_ee(city_id = city_id)
    # clip raster
    clip = world_cover_image.clipToCollection(city_boundary_ee)
    # output name 
    output_file_name = 'data/world_cover_esa/world_cover_'+ city_id+'.tif'
    # export
    geemap.ee_export_image(clip, 
                           filename=output_file_name,
                           #scale =120,
                           region=city_boundary_ee.geometry())


```
